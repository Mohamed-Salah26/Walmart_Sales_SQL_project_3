### EDA & Data Cleaning for Walmart Dataset via Pandas

```python
#import our libraries
import pandas as pd
```

```python
#Reading the file
df = pd.read_csv(r'Walmart.csv', encoding_errors='ignore')
df.head()
```

```python
#Explarotary Data Analysis
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
#Check for duplicate
df.duplicated().sum()
```

```python
#Clear duplicate values
df.drop_duplicates(inplace = True)
df.duplicated().sum()
```

```python
df.shape
```

```python
#check for null values
df.isnull().sum()
```

```python
#We will drop the null values
df.dropna(inplace = True)
df.isnull().sum()
```

```python
# For unit_price column we will try to replace it's type
# First Re check the data types
df.dtypes
```

```python
# df['unit_price'].astype(float)
df['unit_price'] = df['unit_price'].str.replace('$','').astype(float)
df.dtypes
```

```python
df.head()
```

```python
df.info()
```

```python
#Add Profit column
df['Profit'] = (df['unit_price'] * df['quantity']).round(2)
```
